```python
def series_setdiff(s1,s2):
    diff = np.setdiff1d(s1.dropna().unique(), s2.dropna().unique())
    print('Length:', len(diff))
    return diff

def series_intersect(s1,s2):
    intersection = np.intersect1d(s1.dropna().unique(), s2.dropna().unique())
    print('Length:', len(intersection))
    return intersection

def df_strip(df,keep_punc_only=True):
    
    df = df.copy()
    
    for col in df.columns:
        if df[ col ].dtype == 'O':
            if keep_punc_only:
                
                mask = ( df[ col ].str.fullmatch( '[' + punctuation + ']+' ) & df[ col ].notna() )
                only_punc = df.loc[ mask, col ]
                
                df[ col ] = df[ col ].str.strip( ' \t\r\n' + punctuation )
                
                df.update( only_punc )
                
            else:
                df[ col ] = df[ col ].str.strip( ' \t\r\n' + punctuation )
        else:
            continue
            
    return df
```

#### tags

```python
tt = pd.read_csv('./trebi/csv/tags.csv', sep=';')
tt = tt[tt['popularity']<=10]
tt = tt[tt['song_spotify_id'].isin(ts['song_spotify_id'].values)]

tt[['genre','tag']] = df_strip(tt[['genre','tag']])

tt['tag'] = tt['tag'].str.lower()
tt['genre'] = tt['genre'].str.lower()
```

```python
cols = ['tag','popularity']
tt_tag_counts = tt[cols].groupby('tag').count().reset_index().rename(columns={'popularity': 'count'})

my_tags = tt_tag_counts[tt_tag_counts['count']>100]
```

```python
tt = tt.loc[tt['tag'].isin(my_tags['tag'].values)].drop('popularity',axis=1)
tt = tt.drop_duplicates()
```
